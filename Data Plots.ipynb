```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('pair0001.txt').T
plt.scatter(data[0], data[1])
plt.xlabel('Altitude (m)')
plt.ylabel('Temperature ($^\circ$C)')
plt.title('Altitude vs Temperature')
plt.show()
plt1 = plt
print('              Altitude -> Temperature')
print('x Range', min(data[0]), max(data[0]))
print('y Range', min(data[1]), max(data[1]))
```
